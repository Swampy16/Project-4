```python
pollutants = ['CN', 'CO', 'CS2', 'NO2', 'O3', 'SO2']
```

```python
max_pollutant = df[pollutants] = df[pollutants].apply(pd.to_numeric, errors='coerce')
```

```python
max_pollutant['Highest Pollutant Value'] = df[pollutants].max(axis=1)
```

```python
max_pollutant['Highest Pollutant'] = df[pollutants].idxmax(axis=1)
max_pollutant['County'] = df['County']
```

```python
result = max_pollutant[['County', 'Highest Pollutant', 'Highest Pollutant Value']]
print(result)
```

```python
output_file = "pollutant.csv"
result.to_csv(output_file, index=False)
```

```python
parts_per_mil = ['PM10', 'PM2.5']
max_pm = df[parts_per_mil] = df[parts_per_mil].apply(pd.to_numeric, errors='coerce')
max_pm['Highest Parts Per Million Value'] = df[parts_per_mil].max(axis=1)
max_pm['Highest Parts Per Million'] = df[parts_per_mil].idxmax(axis=1)
max_pm['County'] = df['County']
```

```python
result2 = max_pm[['County', 'Highest Parts Per Million Value', 'Highest Parts Per Million']]
print(result2)
```

```python
output_file = 'Parts_per_mil.csv'
result2.to_csv(output_file, index=False)
```

```python
eqi = ['Land_EQI', 'Sociod_EQI', 'Built_EQI', 'Air_EQI', 'Water_EQI']
max_eqi = max_eqi[eqi] = df[eqi].apply(pd.to_numeric, errors='coerce')
max_eqi['Highest EQI Values'] = df[eqi].max(axis=1)
max_eqi['Highest EQI'] = df[eqi].idxmax(axis=1)
max_eqi['County'] = df['County']
```
